```python
need_info = ['LEAS_STATION', 'RTN_STATION']
b = a[need_info]
```

```python
b = b.reset_index(drop=True)
```

```python
# 대여한 자전거정류장 위도경도 채우기
b['대여_X'] = 0
b['대여_Y'] = 0


for i in range(len(b['LEAS_STATION'])):
    for j in range(len(station_info['Station_ID'])):
        if b['LEAS_STATION'].iloc[i] == station_info['Station_ID'].iloc[j] :
            print(i, b['LEAS_STATION'].iloc[i], station_info['Station_ID'].iloc[j])
            b['대여_X'].iloc[i] = station_info['위도'].iloc[j]
            b['대여_Y'].iloc[i] = station_info['경도'].iloc[j]
            break

```

```python
# 반납한 자전거 정류장 위도경도 찍기
b['반납_X'] = 0
b['반납_Y'] = 0


for i in range(len(b['RTN_STATION'])):
    for j in range(len(station_info['Station_ID'])):
        if b['RTN_STATION'].iloc[i] == station_info['Station_ID'].iloc[j] :
            print(i, b['RTN_STATION'].iloc[i], station_info['Station_ID'].iloc[j])
            b['반납_X'].iloc[i] = station_info['위도'].iloc[j]
            b['반납_Y'].iloc[i] = station_info['경도'].iloc[j]
            break
```
